```python
import json
import pandas as pd
import numpy as np
```

```python
with open("../sources/metadata/image_database.json", "r") as data:
    image_database = json.load(data)
```

```python
for image_info in image_database:
    if image_info['folder'] in ['winter_activities', 'summer', 'freedom', 'cake_party']:
        image_info['folder'] = "myPhoto"
```

```python
def flatten_image_data(image_db):
    flattened_data = []
    for item in image_db:
        flat_item = {
            'folder': item.get('folder', ''),
            'date': item.get('date'),
            'path': item.get('path', ''),
            'caption': item.get('caption', ''),
            'faces_label': ', '.join(item.get('faces_label', "")),
            'total_faces': item.get('total_faces', 0),
            'objects_label': ','.join(item.get('objects_label', "")),
            'total_objects': item.get('total_objects', 0),
            'text': item.get('text', ''),
            'cluster_id': item.get('cluster_id', -1),
            'background_class': item.get('background_class', 'None'),
            'color_palette': item.get('color_palette'),
            'feature': item.get('feature', []),
        }
                
        flattened_data.append(flat_item)
    
    return pd.DataFrame(flattened_data)
```

```python
image_db = image_database
df = flatten_image_data(image_db)
df.to_csv('image_database.csv', index=False)
df = pd.read_csv('image_database.csv')
df
```
